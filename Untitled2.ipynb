```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_paths = ["data/movies.csv", "data/movies_metadata.csv", "movies.csv"]
df = None
for p in csv_paths:
    if os.path.exists(p):
        try:
            df = pd.read_csv(p, low_memory=False)
            print("Cargado:", p)
            break
        except Exception as e:
            print("Error leyendo", p, e)
if df is None:
    raise FileNotFoundError("No encontré el CSV en data/. Descarga desde Kaggle y ponlo en data/movies.csv")
```

```python
print("Valores nulos por columna:")
display(df.isnull().sum())

print("\nEstadísticos descriptivos:")
display(df.describe(include='all'))
```

```python
# 2) Vista rápida
print("Filas, columnas:", df.shape)
display(df.head())
display(df.dtypes)

# 3) Chequeo de columnas que normalmente interesan
cols_sugeridas = ['title','release_date','budget','revenue','popularity','vote_average','vote_count','genres','overview','runtime','original_language']
print("Columnas encontradas relevantes:", [c for c in cols_sugeridas if c in df.columns])

# 4) Crear target 'success' según disponibilidad
if 'revenue' in df.columns and df['revenue'].notnull().any():
    df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce').fillna(0)
    revenue_med = df['revenue'].median()
    df['success'] = (df['revenue'] >= revenue_med).astype(int)
    print("Usando 'revenue' como base para success. Mediana:", revenue_med)

elif 'popularity' in df.columns:
    df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce').fillna(df['popularity'].median())
    pop_med = df['popularity'].median()
    df['success'] = (df['popularity'] >= pop_med).astype(int)
    print("Usando 'popularity' como proxy. Mediana:", pop_med)

elif 'vote_average' in df.columns:
    df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce').fillna(df['vote_average'].median())
    vote_med = df['vote_average'].median()
    df['success'] = (df['vote_average'] >= vote_med).astype(int)
    print("Usando 'vote_average' como proxy. Mediana:", vote_med)

else:
    raise ValueError("No encontré columnas útiles para definir success.")

# 5) Feature engineering básica
if 'release_date' in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    df['release_year'] = df['release_date'].dt.year.fillna(0).astype(int)
    df['release_month'] = df['release_date'].dt.month.fillna(0).astype(int)

if 'overview' in df.columns:
    df['overview_len'] = df['overview'].fillna('').apply(lambda s: len(s.split()))

# 6) Selección de features
numeric_feats = []
for c in ['popularity','vote_average','vote_count','overview_len','release_year','release_month']:
    if c in df.columns:
        numeric_feats.append(c)

cat_feats = []
if 'original_language' in df.columns:
    cat_feats.append('original_language')

print("Numeric features:", numeric_feats)
print("Categorical features:", cat_feats)

# 7) Dataset final
model_df = df.dropna(subset=['success']).copy()
X = model_df[numeric_feats + cat_feats]
y = model_df['success']

print("Dataset final listo para entrenar. Dimensiones:", X.shape)
```

```python
# === Imports necesarios para entrenamiento ===
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, StratifiedKFold

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

RANDOM_STATE = 42

```

```python
# 8) Preprocesamiento: ColumnTransformer
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_feats),
    ('cat', categorical_transformer, cat_feats)
], remainder='drop')

# 9) Split train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=RANDOM_STATE
)

# 10) Modelos baseline
models = {
    'logreg': LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    'rf': RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE),
    'knn': KNeighborsClassifier(),
}

results = {}

for name, model in models.items():
    pipe = Pipeline(steps=[('preproc', preprocessor), ('clf', model)])
    pipe.fit(X_train, y_train)

    y_pred = pipe.predict(X_test)
    y_proba = (
        pipe.predict_proba(X_test)[:, 1]
        if hasattr(pipe.named_steps['clf'], 'predict_proba')
        else None
    )

    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall': recall_score(y_test, y_pred, zero_division=0),
        'f1': f1_score(y_test, y_pred, zero_division=0),
        'roc_auc': roc_auc_score(y_test, y_proba) if y_proba is not None else None
    }

    results[name] = metrics
    print(name, metrics)
```

```python
# 11) Fine-tuning con GridSearch para RandomForest

param_grid = {
    'clf__n_estimators': [100, 200, 300],
    'clf__max_depth': [5, 10, 20, None],
    'clf__min_samples_split': [2, 5, 10],
}

rf_pipe = Pipeline([
    ('preproc', preprocessor),
    ('clf', RandomForestClassifier(random_state=RANDOM_STATE))
])

cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=RANDOM_STATE)

gs = GridSearchCV(
    rf_pipe,
    param_grid,
    scoring='roc_auc',
    cv=cv,
    n_jobs=-1,
    verbose=1
)

gs.fit(X_train, y_train)

print("Mejor score AUC (CV):", gs.best_score_)
print("Mejores parámetros:", gs.best_params_)

best_model = gs.best_estimator_
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    roc_curve, precision_recall_curve
)

# === 1. Predicciones ===
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

# === 2. Reporte de clasificación completo ===
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred, digits=4))

# === 3. Métricas adicionales ===
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_proba)

print("\n=== Métricas adicionales ===")
print(f"Accuracy       : {acc:.4f}")
print(f"Precision      : {prec:.4f}")
print(f"Recall         : {rec:.4f}")
print(f"F1-score       : {f1:.4f}")
print(f"ROC-AUC        : {auc:.4f}")

# === 4. Matriz de confusión con análisis ===
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d')
plt.title('Matriz de Confusión')
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()

tn, fp, fn, tp = cm.ravel()

print("\n=== Análisis de la matriz de confusión ===")
print(f"Verdaderos Negativos  (TN): {tn}")
print(f"Falsos Positivos       (FP): {fp}")
print(f"Falsos Negativos       (FN): {fn}")
print(f"Verdaderos Positivos   (TP): {tp}")

# === 5. Curva ROC ===
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label='ROC Curve')
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curva ROC")
plt.legend()
plt.show()

# === 6. Curva Precision-Recall ===
prec_curve, rec_curve, thresh_pr = precision_recall_curve(y_test, y_proba)
plt.figure(figsize=(6,4))
plt.plot(rec_curve, prec_curve)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Curva Precision-Recall")
plt.show()

# === 7. Cálculo del umbral óptimo (Youden's J) ===
youden_j = tpr - fpr
idx = np.argmax(youden_j)
optimal_threshold = thresholds[idx]

print("\n=== Umbral óptimo según Youden J ===")
print(f"Threshold óptimo: {optimal_threshold:.4f}")
print(f"TPR (sensibilidad): {tpr[idx]:.4f}")
print(f"FPR: {fpr[idx]:.4f}")

# === 8. Re-evaluar usando el umbral óptimo ===
y_pred_opt = (y_proba >= optimal_threshold).astype(int)
print("\n=== Reporte con threshold óptimo ===")
print(classification_report(y_test, y_pred_opt, digits=4))
```
